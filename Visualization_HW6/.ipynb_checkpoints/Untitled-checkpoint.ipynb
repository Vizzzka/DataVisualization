```python
import altair as alt
import pandas as pd
```

Проблеми оригінальної візуалізації: важко побачити зріст галузі в регіоні через те що початок зсунутий для всіх галузей окрім першої, важко порівняти зріст конкретної галузі в різних містах, не видно точного відсотка для кожної галузі.

```python
df = pd.read_csv('1.4_02.csv')
```

```python
df.head()
```

```python
lines = alt.Chart(df).mark_line().encode(
    x = alt.X('max(сільське господарство):Q'),
    x2 = alt.X2('min(сільське господарство):Q'),
    y = alt.Y('Регіон:N', sort=alt.Sort(field='сільське господарство', op='max', order='descending')))

points = alt.Chart(df).mark_circle(size = 50).encode(
    x = alt.X('сільське господарство:Q'),
    tooltip = [alt.Tooltip('сільське господарство')],
    y = alt.Y('Регіон:N', sort=alt.Sort(field='сільське господарство', op='max', order='descending')),
    color  = alt.Color('рік:N')
)

chart1 = alt.layer(lines, points).properties(width = 400, height = 300)
```

```python
lines = alt.Chart(df).mark_line().encode(
    x = alt.X('max(промисловість):Q'),
    x2 = alt.X2('min(промисловість):Q'),
    y = alt.Y('Регіон:N', sort=alt.Sort(field='промисловість', op='max', order='descending')))

points = alt.Chart(df).mark_circle(size = 50).encode(
    x = alt.X('промисловість:Q'),
    tooltip = [alt.Tooltip('промисловість')],    
    y = alt.Y('Регіон:N', sort=alt.Sort(field='промисловість', op='max', order='descending')),
    color  = alt.Color('рік:N')
)

chart2 = alt.layer(lines, points).properties(width = 400, height = 300)
```

```python
lines = alt.Chart(df).mark_line().encode(
    x = alt.X('max(будівництво):Q'),
    x2 = alt.X2('min(будівництво):Q'),
    y = alt.Y('Регіон:N', sort=alt.Sort(field='будівництво', op='max', order='descending')))

points = alt.Chart(df).mark_circle(size = 50).encode(
    x = alt.X('будівництво:Q'),
    tooltip = [alt.Tooltip('будівництво')],  
    y = alt.Y('Регіон:N', sort=alt.Sort(field='будівництво', op='max', order='descending')),
    color  = alt.Color('рік:N')
)

chart3 = alt.layer(lines, points).properties(width = 400, height = 300)
```

```python
lines = alt.Chart(df).mark_line().encode(
    x = alt.X('max(послуги):Q'),
    x2 = alt.X2('min(послуги):Q'),
    y = alt.Y('Регіон:N', sort=alt.Sort(field='послуги', op='max', order='descending')))

points = alt.Chart(df).mark_circle(size = 50).encode(
    x = alt.X('послуги:Q'),
    tooltip = [alt.Tooltip('послуги')],
    y = alt.Y('Регіон:N', sort=alt.Sort(field='послуги', op='max', order='descending')),
    color  = alt.Color('рік:N')
)

chart4 = alt.layer(lines, points).properties(width = 400, height = 300)
```

```python
lines = alt.Chart(df).mark_line().encode(
    x = alt.X('max(державне управління):Q'),
    x2 = alt.X2('min(державне управління):Q'),
    y = alt.Y('Регіон:N', sort=alt.Sort(field='державне управління', op='max', order='descending')))

points = alt.Chart(df).mark_circle(size = 50).encode(
    x = alt.X('державне управління:Q'),
    tooltip = [alt.Tooltip('державне управління')],
    y = alt.Y('Регіон:N', sort=alt.Sort(field='державне управління', op='max', order='descending')),
    color  = alt.Color('рік:N')
)

chart5 = alt.layer(lines, points).properties(width = 400, height = 300)
```

```python
alt.vconcat(alt.hconcat(chart3, chart4), alt.hconcat(chart1, chart2), chart5)
```

Я використала гантельковий графік і позначила різні роки різними кольорами. Тепер можно чітко бачити зміни в кожній з галузей для всіх регіонів і порівняти зміни з іншими регіонами. Загалом такий графік читається простіше, ніж згрупований стовпчиковий графік.Також до точок графіку додано підписи з точним значенням відсотку для обох років чого не має на оригінальній візуалізації.
