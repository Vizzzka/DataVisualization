```python
df = pd.read_csv ("GLB.Ts_dSST (1).csv", header=1)
```

```python
df.head()
```

```python
df['mean'] = df[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean(axis=1)
```

```python
average_19th
```

```python
df['mean'] = df['mean'] - average_19th
```

```python
temperature = df[['Year', 'mean']]
```

```python
line = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color='grey').encode(y = alt.Y(field='y', type = 'quantitative'))
```

```python
marked_years_right_df = temperature[temperature['Year'].isin([1944, 2014, 2015, 2017])]
marked_years_left_df = temperature[temperature['Year'].isin([1998, 1904])]
marked_years_bold_df = temperature[temperature['Year'] == 2016]
```

```python
marked_years_right = alt.Chart(marked_years_right_df).mark_text(align='center', dx=22, fontSize=14).encode(x = alt.X(field='Year',
                                                                                            type='quantitative'
                                                                                           ),
                                                                                  y = alt.Y(field='mean',
                                                                                            title='',
                                                                                            type = 'quantitative'),
                                                                                  text='Year:Q',
                                                                                 )
```

```python
marked_years_left = alt.Chart(marked_years_left_df).mark_text(align='center', dx=-22,fontSize=14).encode(x = alt.X(field='Year',
                                                                                            type='quantitative'
                                                                                           ),
                                                                                  y = alt.Y(field='mean',
                                                                                            title='',
                                                                                            type = 'quantitative'),
                                                                                  text='Year:Q',
                                                                                 )
```

```python
marked_years_bold = alt.Chart(marked_years_bold_df).mark_text(align='center', dx=22,fontWeight='bold',fontSize=14).encode(x = alt.X(field='Year',
                                                                                            type='quantitative'
                                                                                           ),
                                                                                  y = alt.Y(field='mean',
                                                                                            title='',
                                                                                            type = 'quantitative'),
                                                                                  text='Year:Q',
                                                                                 )
```

```python
base = alt.Chart(temperature).mark_point(color='black', fill='#F26C08', strokeWidth=1, size=70).encode(
    x = alt.X(field='Year',
              title='',
              type = 'quantitative',
              scale = alt.Scale(zero=False),
              axis = alt.Axis(tickCount = 14, format = '<4d')
             ),
    y = alt.Y(field='mean',
              title='',
              type = 'quantitative',
              axis = alt.Axis(labelExpr= 'datum.label+ "\u00b0"', tickCount=6)
             ),
        
)
```

```python
title = alt.Chart(temperature).mark_text(fontSize=15, align='left', lineBreak='\n', dy=20, dx=20).encode(text=alt.value("Annual Global Surface Temperature,\n Relative to Late 19th Century Average"), x=alt.value(0), y=alt.value(0))
```

```python
line_text = alt.Chart(pd.DataFrame({'y': [0]})).mark_text(fontSize=12, align='right', lineBreak='\n', dy=-30, dx=445).encode(text=alt.value("HOTTER THAN THE\n 1880-1899 AVERAGE"), y = alt.Y(field='y', type = 'quantitative'))
```

```python
line_text_down = alt.Chart(pd.DataFrame({'y': [0]})).mark_text(fontSize=12, align='right', lineBreak='\n', dy=15, dx=445).encode(text=alt.value("COLDER"), y = alt.Y(field='y', type = 'quantitative'))
```

```python
alt.layer(line,
          base,
          title,
          line_text,
          line_text_down,
          marked_years_right,
          marked_years_left,
          marked_years_bold).properties(width=900,
                                        height=500,
                                        padding = 25,
                                        background= '#F9F9F9'
                                       ).configure_axis(labelFontSize=12.5,
                                                        labelColor='grey'
                                                       )
```
